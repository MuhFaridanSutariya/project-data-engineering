### Import libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### Pagination + multiple page

```python
hotel_name = []
hotel_original_price = []
hotel_price_after_discount = []
tax = []
rating = []
location = []



for i in range(0, 100):
    # url
    website = 'https://m.traveloka.com/en-id/hotel/search/GEO.102746.1.08-10-2022.09-10-2022?&skip='+ str(i) + '-prev&sid=383c3561-fe37-46d1-a5e2-213c477c9be6.1746093390140821347'
    
    # request
    response = requests.get(website)
    
    # soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # results
    results = soup.find_all("div", {"class":"Dbn77"})

    
    
    # loop results
    for result in results:
        #hotel name
        try:
            hotel_name.append(result.find("div", {"class":"_16ZGw"}).get_text())
        except:
            hotel_name.append('n/a')

        #original price hotel
        try:
            hotel_original_price.append(result.find("div", {"class":"tvat-originalPrice"}).get_text())
        except:
            hotel_original_price.append('n/a')

        #price hotel after discount
        try:
            hotel_price_after_discount.append(result.find("span", {"class":"tvat-hotelPrice"}).get_text())
        except:
            hotel_price_after_discount.append('n/a')

        #tax
        try:
            tax.append(result.find("div", {"class":"_15h-n"}).get_text())
        except:
            tax.append('n/a')

        #rating
        try:
            rating.append(result.find("span", {"class":"tvat-ratingScore"}).get_text())
        except:
            rating.append('n/a')

        try:
            location.append(result.find(text=lambda text: text and ", Bali" in text))
        except:
            location.append('n/a')
```

```python
product_overview = pd.DataFrame({'Hotel Name':hotel_name, 'Original price':hotel_original_price, 
                                'Price after discount':hotel_price_after_discount, 
                                'Tax':tax, 'Tax':tax, 'Rating':rating, 'location':location})
```
